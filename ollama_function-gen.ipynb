```python
import requests

def ollama_request(prompt):
    # Send a request to the local Ollama API
    url = "http://localhost:11434/api/chat"  # Update if using a different host/port
    payload = {
        #"model": "codegemma:2b",
        "model": "llama3:8b",
        "prompt": prompt
    }

    try:
        response = requests.post(url, json=payload)
        response_data = response.json()

        # Print full response for debugging
        print("Full Response:", response_data)

        # Extract and return the 'response' field
        generated_function = response_data.get("response", "")
        if not generated_function:
            raise ValueError("Empty response. Check the prompt or model setup.")
        return generated_function

    except Exception as e:
        print("Error:", e)
        return None

# Test with a basic prompt
prompt = "Write a Python function that multiplies two numbers. Include a detailed docstring. Only return the code."
function_code = ollama_request(prompt)
if function_code:
    print("Generated Function:\n", function_code)
```
